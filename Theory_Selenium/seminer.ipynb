# Webdriver Waits

<font color='red'>Why we need to Waits when we user webdriver ?</font> 

1.   <font color='orange'>Implicit wait  </font>: when we use "implicit wait" after run test script, it will be ask the webdriver to pause for amount of time initially set ,before throwing an exception that says something like this: “No Such Element Exception”.

<font color='orrange'>syntax    </font>: driver.implicitly_wait(<font color ="yellow">**TimeOut** </font>)

```python
# Example code 
driver.get("D:/Study/LT KHDL/seminer/index.html")

driver.implicitly_wait(10)
#test tag id have in html
my_dynamic_element = driver.find_element(By.ID, "username")
print(my_dynamic_element)
```

2.   <font color='orange'>Expected Wait  </font> : Unlike Implicit Wait, when using expected wait it has been enclose with any conditions, it's mean instead of waiting for a pre-set amount of time ,here they will wait for specific conditions or check maximun time out before output an exception something like "Element Not Visible Exception"

<font color ="orrange">syntax   </font> : WebDriverWait(driver, <font color ="yellow">**TimeOut** </font>).until(<font color ="yellow">some_condition </font>)\
`*` some conditons :
* alert is present
* element exists
* element is visible
* title contains
* title is
* element staleness
* visible text

```python
#Example Code

#check 
driver.get("D:/Study/LT KHDL/seminer/index.html")
delay = 3 # seconds
try:
    myElem = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.ID, 'usernatme')))
    print("page is ready!")
except TimeoutException:
    print ("Loading took too much time!")
```

3.   <font color='orange'>Fluent wait  </font> : Use Fluent Wait in situations like , sometimes we encounter elements that take in a few minutes to load,but other times it takes more time than until the times limit . Fluent wait will constant search until the elements is found or until times out 

<font color ="orrange">syntax   </font> : WebDriverWait(driver, <font color ="yellow">**TimeOut** </font>,<font color ="yellow">**Poll Frequency** </font> , ignored_exceptions=[ElementNotVisibleException, ElementNotSelectableException])

```python
#example code

#check true

delay = 5
wait = WebDriverWait(driver, delay, poll_frequency=1, ignored_exceptions=["ElementNotVisibleException", "ElementNotSelectableException"])
element = wait.until(EC.element_to_be_clickable((By.ID, "username")))
#print(element)
```

- <font color="lightyellow"><h5>Prompt  : </h5></font>Prompts are similar to confirm boxes, except they also include a text input. Similar to working with form elements, you can use WebDriver’s send keys to fill in a response. This will completely replace the placeholder text. Pressing the cancel button will not submit any text

```python
from selenium.common.exceptions import UnexpectedAlertPresentException
while True:
    try:
        driver.execute_script("var a = prompt('Type anything', 'hello');document.body.setAttribute('user-manual-input', a)")
        time.sleep(5)  # must 
        print(driver.find_element(By.TAG_NAME,"body").get_attribute('user-manual-input')) # get the text
        break

    except (UnexpectedAlertPresentException):
        pass
```

3.  <font color='orange'>Cookies  </font> :

- <font color="lightyellow"><h5>Add Cookie  : </h5></font>It is used to add a cookie to the current browsing context. Add Cookie only accepts a set of defined serializable JSON object. Here is the link to the list of accepted JSON key values

```python
# Adds the cookie into current browser context
driver.add_cookie({"name": "key", "value": "value"})
```

- <font color="lightyellow"><h5>Get Named Cookie : </h5></font>It returns the serialized cookie data matching with the cookie name among all associated cookies.

```python
# Adds the cookie into current browser context
driver.add_cookie({"name": "foo", "value": "bar"})
# Get cookie details with named cookie 'foo'
print(driver.get_cookie("foo"))
```

- <font color="lightyellow"><h5>Get All Cookies  : </h5></font>It returns a ‘successful serialized cookie data’ for current browsing context. If browser is no longer available it returns error.

```python
driver.add_cookie({"name": "test1", "value": "cookie1"})
driver.add_cookie({"name": "test2", "value": "cookie2"})

# Get all available cookies
print(driver.get_cookies())
```

- <font color="lightyellow"><h5>Delete Cookie  : </h5></font>It deletes the cookie data matching with the provided cookie name.

```python
driver.add_cookie({"name": "test1", "value": "cookie1"})
driver.add_cookie({"name": "test2", "value": "cookie2"})


# Delete a cookie with name 'test1'
driver.delete_cookie("test1")

#  Deletes all cookies
driver.delete_all_cookies()
```
